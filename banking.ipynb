```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Med','High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins, labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

## Univariate Analysis

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data = df,x=predictor)
```

## Bivariate Analysis

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data = df,x=predictor,hue = 'Nationality')
```

```python
# Histplot of value counts for different Occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.Figure(figsize=(8,4))
    sns.histplot(df[col],color="Red")
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# Numerical Analysis
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

#Univariate analysis and visualization
plt.Figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    plt.tight_layout()
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

# HeatMaps


```python
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

coorelation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,14))
sns.heatmap(coorelation_matrix , annot=True, cmap='crest', fmt=".2f")
plt.title("Coorelation Matrix")
plt.show()
```
